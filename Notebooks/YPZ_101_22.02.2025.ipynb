```python
# Bir e ticaret firmasında her ürünün kaç adet satıldığını gösteren bir veri seti mevcuttur.
data = {
    'urun': ['Telefon','Bilgisayar','Telefon','Telefon','Bilgisayar','Kulaklık','Kulaklık'],
    'adet': [120, 5, 30, 10,25,41,47]
}

df = pd.DataFrame(data)
toplam_satis = df.groupby('urun')['adet'].sum()
toplam_satis
```

```python
# Bir mağazanın satış verisinde her şehirdeki ortalama satış miktarını hesapla
data = {
    'bolge': ['İstanbul','İstanbul','Ankara','Ankara','İzmir','İzmir'],
    'fiyat': [100,120,140,180,110,96]
}

df = pd.DataFrame(data)
ortalama = df.groupby('bolge')['fiyat'].mean()
ortalama
```

```python
# Müşterileri yaş gruplarına göre ayırarak ilgili yaş grubunun ortalama harcamalrını hesaplayalım.
kisiler ={
    'yas': np.random.randint(20,60,25),
    'harcama': np.random.randint(300,8000,25)
}
dF = pd.DataFrame(kisiler)
dF
```

```python
#yaş aralıkları oluştur
# 20-30   -> 3500 
yas_grubu = pd.cut(dF['yas'], bins=[19,30,40,50,60], labels=['20-30','30-40','40-50','50-60'])
yas_grubu
```

```python
# ortalama harcamayı hesapla 
ortalama_harcama = dF.groupby(yas_grubu)['harcama'].mean()
ortalama_harcama
```

```python
data = {
    'musteri_id':[1,2,3,4,5,6,7],
    'isim':['Buse','Ömer','Cavit','Hüseyin','Berker','Gizem','Mihriban'],
    'gelir': [85000,25000,250000,100000,150000,100000,135000],
    'gider':[35000, 35000, 36000, 45000,87000, 3000, 1000]
}

df = pd.DataFrame(data)
df
```

```python
#aylık net gelir
df['Aylık Gelir'] = df['gelir'] - df['gider']
df
```

```python
df['Kredi Riski'] = df['Aylık Gelir'].apply(lambda x:'Düşük' if x<75000 else 'Yüksek')
df
```

```python
df.to_excel('Musteri_Kredi_Riski.xlsx')
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,10,1000)
y = np.tan(x)

plt.plot(x,y)
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,10,1000)
y = np.tan(x)

plt.plot(x,y,'r')
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,10,40)
y = np.sin(x)

# yeşil ve yuvarlak grafik.
plt.plot(x,y,'go')
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.show()
```

```python
# kırmızı ve üçgen şeklinde grafik
plt.figure(figsize=(10,5))
x = np.linspace(0,10,40)
y = np.sin(x)


plt.plot(x,y,'r^')
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.show()
```

```python
# ters üçgen 
plt.figure(figsize=(10,5))
x = np.linspace(0,10,40)
y = np.sin(x)
plt.plot(x,y,'rv')
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.arange(1,7)
y1 = np.power(x,2)
y2 = np.power(x,3)

plt.plot(x,y1, 'b-',label = '$y1 = x^2$')
plt.plot(x,y2, 'r-', label = '$y2 = x^3$')
plt.tight_layout()
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.legend()
plt.show()
```

```python
# Hem sinüs hemde cosinüs değerlerini alan grafiği çizdiriniz.
plt.figure(figsize=(10,6))
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), label='Sin(x)')
plt.plot(x, np.cos(x), label = 'Cos(x)')
plt.xlabel('X', fontsize = 18)
plt.ylabel('Y', fontsize=18)
plt.title('Sin(x) & Cos(X)', fontsize=16)
plt.legend(loc="lower left")
plt.show()
```
