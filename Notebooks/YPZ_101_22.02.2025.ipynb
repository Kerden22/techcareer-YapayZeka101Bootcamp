```python
# Bir e ticaret firmasÄ±nda her Ã¼rÃ¼nÃ¼n kaÃ§ adet satÄ±ldÄ±ÄÄ±nÄ± gÃ¶steren bir veri seti mevcuttur.
data = {
    'urun': ['Telefon','Bilgisayar','Telefon','Telefon','Bilgisayar','KulaklÄ±k','KulaklÄ±k'],
    'adet': [120, 5, 30, 10,25,41,47]
}

df = pd.DataFrame(data)
toplam_satis = df.groupby('urun')['adet'].sum()
toplam_satis
```

```python
# Bir maÄazanÄ±n satÄ±Å verisinde her Åehirdeki ortalama satÄ±Å miktarÄ±nÄ± hesapla
data = {
    'bolge': ['Ä°stanbul','Ä°stanbul','Ankara','Ankara','Ä°zmir','Ä°zmir'],
    'fiyat': [100,120,140,180,110,96]
}

df = pd.DataFrame(data)
ortalama = df.groupby('bolge')['fiyat'].mean()
ortalama
```

```python
# MÃ¼Återileri yaÅ gruplarÄ±na gÃ¶re ayÄ±rarak ilgili yaÅ grubunun ortalama harcamalrÄ±nÄ± hesaplayalÄ±m.
kisiler ={
    'yas': np.random.randint(20,60,25),
    'harcama': np.random.randint(300,8000,25)
}
dF = pd.DataFrame(kisiler)
dF
```

```python
#yaÅ aralÄ±klarÄ± oluÅtur
# 20-30   -> 3500 
yas_grubu = pd.cut(dF['yas'], bins=[19,30,40,50,60], labels=['20-30','30-40','40-50','50-60'])
yas_grubu
```

```python
# ortalama harcamayÄ± hesapla 
ortalama_harcama = dF.groupby(yas_grubu)['harcama'].mean()
ortalama_harcama
```

```python
data = {
    'musteri_id':[1,2,3,4,5,6,7],
    'isim':['Buse','Ãmer','Cavit','HÃ¼seyin','Berker','Gizem','Mihriban'],
    'gelir': [85000,25000,250000,100000,150000,100000,135000],
    'gider':[35000, 35000, 36000, 45000,87000, 3000, 1000]
}

df = pd.DataFrame(data)
df
```

```python
#aylÄ±k net gelir
df['AylÄ±k Gelir'] = df['gelir'] - df['gider']
df
```

```python
df['Kredi Riski'] = df['AylÄ±k Gelir'].apply(lambda x:'DÃ¼ÅÃ¼k' if x<75000 else 'YÃ¼ksek')
df
```

```python
plt.figure(figsize=(10,5))
x = np.arange(1,7)
y1 = np.power(x,2)
y2 = np.power(x,3)

plt.plot(x,y1, 'b-',label = '$y1 = x^2$')
plt.plot(x,y2, 'r-', label = '$y2 = x^3$')
plt.tight_layout()
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.legend()
plt.show()
```

```python
# Hem sinÃ¼s hemde cosinÃ¼s deÄerlerini alan grafiÄi Ã§izdiriniz.
plt.figure(figsize=(10,6))
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), label='Sin(x)')
plt.plot(x, np.cos(x), label = 'Cos(x)')
plt.xlabel('X', fontsize = 18)
plt.ylabel('Y', fontsize=18)
plt.title('Sin(x) & Cos(X)', fontsize=16)
plt.legend(loc="lower left")
plt.show()
```
