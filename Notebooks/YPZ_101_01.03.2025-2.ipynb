```python
X = np.arange(1,11)
y = np.array([1,4,9,16,25,36,49,64,81,100])
```

```python
plt.scatter(X,y, color='blue')
plt.xlabel('Reklam Bütçesi')
plt.ylabel('Gelir')
plt.show()
```

```python
poly_degre = 2
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X.reshape(-1,1))
X_poly
```

```python
model = LinearRegression()
model.fit(X_poly,y)

y_pred = model.predict(X_poly)
```

```python
y_pred
```

```python
plt.scatter(X,y, color="blue",label="Gerçek Veriler")
plt.plot(X,y_pred, color="red", label="Tahminler")
plt.xlabel('Reklam Bütçesi')
plt.ylabel('Gelir')
plt.legend()
plt.show()
```

```python
new_budget = np.array([[12]])
new_budget_poly = poly_features.transform(new_budget)
predicted_gelir = model.predict(new_budget_poly)

print(f'12k Tl reklam bütçesi için tahmin edilen gelir : {predicted_gelir[0]}k Tl')
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df.head(9)
```

```python
df['target'] = cancer.target
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
X.head()
```

```python
y.head(25)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)

accuracy = accuracy_score(y_test,pred)
print(f'Modelin kesinliği : {accuracy}')

cm = confusion_matrix(y_test,pred)
print(cm)
```

```python
plt.bar(['Kesinlik'], [accuracy*100])
plt.ylim(0,100)
plt.ylabel('Accuracy %')
plt.title('Model Doğruluğu')
plt.show()
```

```python
train_size = len(X_train)
test_size = len(X_test)

plt.bar(['Eğitim Verisi','Test Verisi'], [train_size,test_size])
plt.show()
```

```python
# Elimizdeki Verilere göre ilgili kişinin restorana veya cafe ye gidip gitmeme durumunu Decision Tree ile 
# kurgulayınız.
data = {
    'Hava Durumu': ['Güneşli','Güneşli','Bulutlu','Yağmurlu','Yağmurlu','Bulutlu','Güneşli'],
    'Sıcaklık':['Sıcak','Sıcak','Sıcak','Ilık','Soğuk','Ilık','Ilık'],
    'Rüzgar': ['Rüzgarsız','Rüzgarlı','Rüzgarsız','Rüzgarlı','Rüzgarlı','Rüzgarlı','Rüzgarsız'],
    'Olumlu': ['E','E','E','H','H','H','E']
}

df = pd.DataFrame(data)
df
```

```python
# Hava Durumu için Güneşli:0, Bulutlu:1, Yağmurlu:2
# Sıcaklık için Sıcak:0 , Ilık:1, Soğuk: 2 
# Rüzgar : Rüzgarsız:0, Rüzgarlı:1
# Olumlu:  H:0, E:1
df['Hava Durumu'] = df['Hava Durumu'].map({'Güneşli':0, 'Bulutlu':1, 'Yağmurlu':2})
df['Sıcaklık'] = df['Sıcaklık'].map({'Sıcak':0,'Ilık':1,'Soğuk':2})
df['Rüzgar'] = df['Rüzgar'].map({'Rüzgarsız':0, 'Rüzgarlı':1})
df['Olumlu'] = df['Olumlu'].map({'H':0,'E':1})

df
```

```python
X = df.drop('Olumlu',axis=1)
y = df['Olumlu']
```

```python
X
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
hava_durumu = input('Lütfen Hava Durumunu giriniz: ')
sicaklik = input('Lütfen Sıcaklık giriniz: ')
ruzgar = input('Lütfen Rüzgar Durumunu giriniz: ')
new_data = pd.DataFrame({'Hava Durumu':[hava_durumu], 'Sıcaklık':[sicaklik], 'Rüzgar':[ruzgar]})
new_data['Hava Durumu'] = new_data['Hava Durumu'].map({'Güneşli':0, 'Bulutlu':1, 'Yağmurlu':2})
new_data['Sıcaklık'] = new_data['Sıcaklık'].map({'Sıcak':0,'Ilık':1,'Soğuk':2})
new_data['Rüzgar'] = new_data['Rüzgar'].map({'Rüzgarsız':0, 'Rüzgarlı':1})
tahmin = model.predict(new_data)

if tahmin[0]==1:
    print('Bugünün Hava Şatları gayet iyi dışarı çıkabilirsiniz.')
else:
    print('Bugünün Hava Şatları kötü gitmemenizi tavsiye ederim')
```
