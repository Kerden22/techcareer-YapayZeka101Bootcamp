```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x), label = "Sin(X)")
plt.plot(x,np.cos(x), label = 'Cos(X)')
plt.fill_between(x, 0 , np.sin(x))
plt.fill_between(x,0, np.cos(x))
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 'lower left')
plt.show()
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(1,2,1)
plt.title('Sin')
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.title('Cos')
plt.plot(x,y2)
plt.show()
```

```python
plt.figure(figsize=(12,6))
x= np.arange(1,11)
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.subplot(2,1,1)
plt.title('Kare fonksiyon')
plt.plot(x,y1, 'b-')
plt.subplot(2,1,2)
plt.title('Küp fonksiyon')
plt.plot(x,y2, 'r-')
plt.show()
```

```python
x = np.arange(-50,51)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.tan(x)
y4 = np.exp(x)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16,8))


ax[0,0].plot(x, y1, color="orange")
ax[0,0].set_title('Kare fonksiyon')
ax[0,0].set_xlabel('X',fontsize = 10)
ax[0,0].set_ylabel('Y',fontsize = 10)

ax[0,1].plot(x,y2, color = 'red')
ax[0,1].set_title('Küp fonksiyon')
ax[0,1].set_xlabel('X', fontsize=10)
ax[0,1].set_ylabel('Y',fontsize=10)


ax[1,0].plot(x,y3, color = 'blue')
ax[1,0].set_title('Tanjant fonksiyon')
ax[1,0].set_xlabel('X', fontsize=10)
ax[1,0].set_ylabel('Y',fontsize=10)


ax[1,1].plot(x,y4, color = 'purple')
ax[1,1].set_title('Exponansiyel fonksiyon')
ax[1,1].set_xlabel('X', fontsize=10)
ax[1,1].set_ylabel('Y',fontsize=10)

plt.show()
```

```python
# Bar grafik
def f(x):
    return 2*x +7
x = np.arange(1,10)
y = f(x)
plt.bar(x,y)
plt.show()
```

```python
x = np.arange(1,10)
score = np.arange(20,110,10)
plt.figure(figsize=(9,6))
plt.bar(x,score, color = '#f95959')
plt.show()
```

```python
def f(x):
    return x*2 + 5


def g(x):
    return 3*x + 5

x1 = np.array([1,3,5,7])
x2 = np.array([2,4,6,8])
y1 = f(x1)
y2 = g(x2)

plt.bar(x1,y1, label = '2x + 5', color = '#f95959')
plt.bar(x2,y2, label = '3x + 5', color = '#455d7a')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# yan grafik
x1 = np.array([1,3,5,7])
x2 = np.array([2,4,6,8])
y1 = f(x1)
y2 = g(x2)

plt.barh(x1,y1, label = '2x + 5', color = '#f95959')
plt.barh(x2,y2, label = '3x + 5', color = '#455d7a')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Pasta Grafiği 
sales = [45, 35,20,10,65]
models = ['Samsung','Iphone','Huawei','Xiaomi','Nokia']
colors = ['#222831','#393e46','#f96d00','#f2f2f2','#00bbf0']

plt.pie(sales, labels=models, colors= colors)
plt.show()
```

```python
plt.figure(figsize=(8,8))
hasta_sayisi = [150,250,350,450]
total = np.sum(hasta_sayisi)
derece = ['Low','Medium','High','Critical']
renkler = ['#ffebbb','#f1b963','#f96d00','#f2f2f2','#00bbf0']
def deger_goster(x):
    return '{:.2f}%\n({:.1f})'.format(x, total*x/100)
plt.pie(hasta_sayisi, labels= derece, colors=renkler,autopct=deger_goster )
# yüzdelik = (low / toplam_hasta_sayisi) *100
plt.show()
```
